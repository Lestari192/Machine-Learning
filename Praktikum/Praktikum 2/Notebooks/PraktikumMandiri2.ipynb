```
import pandas as pd

# Read the CSV file with a comma delimiter
df = pd.read_csv('/content/drive/MyDrive/MachineLearning/Praktikum2/Data/day.csv')

#cetak header data(5 baris data) dari file
df.head()
```

```
# Total Rows
total_rows = len(df)
print(total_rows)
```

```
# Calculate Size
train_size = int(0.8 * total_rows)
test_size = int(0.2 * total_rows)
print(f"Training Size (80%): {train_size} baris")
print(f"Testing Size (20%): {test_size} baris")
```

```
# Split Data
train_df = df[:train_size]
test_df = df[:train_size]
print("Training Set:", len(train_df))
display(train_df.head())

print("\nTesting Set:", len(test_df))
display(test_df.head())
```

```
# Further Split Training Into Training And Validation (10% of training)
val_size = int(0.1 * len(train_df))
val_df = train_df[:val_size]
train_df = train_df[val_size:]
print("Ukuran Data Validation (10% dari Training):", val_size)
print("Ukuran Data Training Baru (Setelah dipotong Validation):", len(train_df))

print("\nValidation Set (5 baris teratas):")
display(val_df.head())

print("\nTraining Set Baru (5 baris teratas):")
display(train_df.head())
```

```
# Display Result
print("\nTraining Set:", len(train_df))
display(train_df.head())

print("\nValidation Set:", len(val_df))
display(val_df.head())

print("\nTesting Set:", len(test_df))
display(test_df.head())
```
